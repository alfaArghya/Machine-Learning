```python
df = pd.DataFrame(d)
df
```

```python
pd.DataFrame(d, index=["d", "b", "a","z"])
```

```python
pd.DataFrame(d, index=["d", "b", "a","z"], columns = ["two", "three", "one", "ten"] )
```
